Toutes les feuilles dans ce fichier sont en meme format. Donc on utilise touts les étapes précent pour créer une fonction afin de faciliter la démarche de la préparation des données.

### Fonction pour la separation des données

```python
def separation_df(fichier, sheet):
    temp = fichier.parse(sheet)
    temp = pd.DataFrame(temp)
    temp.drop(labels=[0,1,2,5],inplace=True)
    temp = temp.iloc[:-3 , :]
    temp.dropna(axis=1,how='any',inplace=True)
    
    labels = temp.iloc[0]
    temp.drop(labels=[3,4], inplace = True)
    
    temp.columns = labels
    temp.rename(columns = {temp.columns[0]:"année"},inplace = True)
    
    temp.reset_index()
    return temp
```

On utilise un boucle pour appliquer cette fonction au but d'obtenir un dataframe concatné avec touts les données

```python
#Crétaion de variable pour le résultat final

fin_emp_cat = pd.DataFrame()
for i in range (0,len(sheets_emploi_cat)):
    
    #Utilisation de la fonction
    temp = separation_df(file_emploi_cat,sheets_emploi_cat[i])
    
    #Ajouté les type de données (age et sexe)
    temp['new'] = sheets_emploi_cat[i]
    
    #Concatener avec les données précedent
    fin_emp_cat = pd.concat([fin_emp_cat,temp])
    
fin_emp_cat
```

Donc on va convertir tout les étapes pour une fonction qu'on puisse isolater des données au csv

```python
file_path = '../Raw_data/departement/population/estim-pop-dep-sexe-aq-1975-2021.xlsx'

def convertir_exl_csv(inputpath,sheet,nrows,usecols,skiprows):
    
    temp_df = pd.read_excel(inputpath,sheet_name=sheet,nrows=nrows,usecols=usecols,skiprows=skiprows)
    temp_df.rename(columns = {temp_df.columns[0]:"code_postal"},inplace=True)
    return temp_df
    
convertir_exl_csv(file_path,sheets_pop_dep[1],96, "A,X:AQ", 4)
```

Les formats de feuilles sont differentes avant 1999 et après 1999. Donc on a créer une simple finction au-dessus afin d'isoler la partie utile.
